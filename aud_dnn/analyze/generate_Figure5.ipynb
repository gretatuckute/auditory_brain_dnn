## Component data (in-house models, trained), barplots (Figure 5A)

```python
target = 'NH2015comp'
permuted_network = 'False' 

# Define the order of models  (this order is based on the performance of models for NH2015 in Figure 2A)
sort_flag_manual = ['ResNet50multitask', 'AST', 'VGGish', 'ResNet50audioset', 'Kell2018multitask',
                     'ResNet50word', 'ResNet50speaker', 'Kell2018speaker', 'Kell2018word',
                     'Kell2018audioset', 'Kell2018music', 'ResNet50music', 'S2T', 'wav2vec',
                     'sepformer', 'DS2', 'ZeroSpeech2020', 'DCASE2020', 'metricGAN']

barplot_components_across_models(source_models=sort_flag_manual,
                                 target=target,
                                 randnetw=permuted_network,
                                 value_of_interest='median_r2_test',
                                 yerr_type='median_r2_test_sem_over_it',
                                 save=SAVEDIR,
                                 add_savestr=f'',
                                 include_spectemp=True,
                                 sort_by=sort_flag_manual,
                                 add_in_spacing_bar=False,
                                 ylim=[0,1],
                                 box_aspect=0.8)
```
